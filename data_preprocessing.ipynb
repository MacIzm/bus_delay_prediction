### Splitting Occured_On into two columns - Date and time

```python
date_time = subsampled_data["Occurred_On"].str.split(" ", n = 1, expand = True)
subsampled_data["date"] = date_time[0]
subsampled_data["time"] = date_time[1]
subsampled_data.drop(columns =["Occurred_On"], inplace = True)
subsampled_data
```

### Converting School_Year, Run_Type, Reason, Boro, Breakdown_or_Running_Late to categorical format

```python
subsampled_data['School_Year'] = pd.Categorical(subsampled_data.School_Year)
subsampled_data['Run_Type'] = pd.Categorical(subsampled_data.Run_Type)
subsampled_data['Reason'] = pd.Categorical(subsampled_data.Reason)
subsampled_data['Boro'] = pd.Categorical(subsampled_data.Boro)
subsampled_data
```

```python
subsampled_data = pd.concat([subsampled_data, 
                             pd.get_dummies(subsampled_data['Run_Type'], prefix='Run_Type')],axis=1)
subsampled_data.drop("Run_Type", axis = 1, inplace=True)

subsampled_data = pd.concat([subsampled_data, 
                             pd.get_dummies(subsampled_data['Boro'], prefix='Boro')],axis=1)
subsampled_data.drop("Boro", axis = 1, inplace=True)

subsampled_data = pd.concat([subsampled_data, 
                             pd.get_dummies(subsampled_data['Reason'], prefix='Reason')],axis=1)
subsampled_data.drop("Reason", axis = 1, inplace=True)
```

### Creating hash values for Bus_No, Route_Number, Schools_Serviced, and Bus_Company_Name

```python
bus_no_unique = subsampled_data['Bus_No'].unique()
bus_no_dict = { bus_no_unique[i]:i for i in range(0, len(bus_no_unique) ) }
    
route_no_unique = subsampled_data['Route_Number'].unique()
route_no_dict = { route_no_unique[i]:i for i in range(0, len(route_no_unique) ) }
    
school_no_unique = subsampled_data['Schools_Serviced'].unique()
school_no_dict = { school_no_unique[i]:i for i in range(0, len(school_no_unique) ) }
    
bus_company_unique = subsampled_data['Bus_Company_Name'].unique()
bus_company_dict = { bus_company_unique[i]:i for i in range(0, len(bus_company_unique) ) }
```

```python
def to_int(val, int_dict):
    return int_dict[val]

subsampled_data.Bus_No = subsampled_data.Bus_No.apply(to_int, args = [bus_no_dict])
subsampled_data.Route_Number = subsampled_data.Route_Number.apply(to_int, args = [route_no_dict])
subsampled_data.Schools_Serviced = subsampled_data.Schools_Serviced.apply(to_int, args = [school_no_dict])
subsampled_data.Bus_Company_Name = subsampled_data.Bus_Company_Name.apply(to_int, args = [bus_company_dict])
subsampled_data
```

```python
subsampled_data.drop("date", axis = 1, inplace=True)
subsampled_data.drop("time", axis = 1, inplace=True)
```

### Splitting data into training and testing datasets
All the school years of 2018-2019 are used as the testing data points, and rest as the training data points

```python
test = subsampled_data[subsampled_data["School_Year"] == "2018-2019"]
train = subsampled_data.drop(subsampled_data[subsampled_data["School_Year"] == "2018-2019"].index)

train.drop("School_Year", axis = 1, inplace=True)
test.drop("School_Year", axis = 1, inplace=True)
```

### Saving the preprocessed data to a csv file
