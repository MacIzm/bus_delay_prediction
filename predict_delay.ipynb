# Predicting the bus delay 

### Reading the preprocessed datasets for training and testing

```python
import pandas as pd
import matplotlib.pyplot as plt
import collections

train_path = '../data/train.csv'
test_path = '../data/test.csv'
```

```python
train = pd.read_csv(train_path, sep = '\t')
test = pd.read_csv(test_path, sep='\t')
train
```

```python
train_y = []
for index, row in train.iterrows():
        train_y.append(int(row['How_Long_Delayed']))
train_x = train.drop("How_Long_Delayed", axis = 1)
```

```python
test_y = []
for index, row in test.iterrows():
        test_y.append(int(row['How_Long_Delayed']))
test_x = test.drop("How_Long_Delayed", axis = 1)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg = LinearRegression().fit(train_x, train_y)
test_y_ = reg.predict(test_x)

```

```python
import numpy as np

np.sqrt(mean_squared_error(test_y_, test_y)), np.mean(test_y)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

regr_1 = DecisionTreeRegressor(max_depth=2)
regr_1.fit(train_x, train_y)

# Predict
y_1 = regr_1.predict(test_x)

np.sqrt(mean_squared_error(y_1, test_y)), np.mean(test_y)
```

## Polynomial regression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial', max_iter = 200).fit(train_x, train_y)
test_y_ = clf.predict(test_x)

np.sqrt(mean_squared_error(test_y_, test_y)), np.mean(test_y)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
regr.fit(train_x, train_y)

test_y_RF = regr.predict(test_x)

np.sqrt(mean_squared_error(test_y_RF, test_y)), np.mean(test_y)
```

## Artificial Neural Network

```python
from sklearn.neural_network import MLPRegressor
reg = MLPRegressor(hidden_layer_sizes=(200,150),  activation='relu', solver='adam',    alpha=0.001,batch_size='auto',
               learning_rate='constant', learning_rate_init=0.01, power_t=0.5, max_iter=1000, shuffle=True,
               random_state=None, tol=0.0001, verbose=False, warm_start=False, momentum=0.9,
               nesterovs_momentum=True, early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999,
               epsilon=1e-08)

reg = reg.fit(train_x, train_y)

test_y_ann = reg.predict(test_x)

np.sqrt(mean_squared_error(test_y_ann, test_y)), np.mean(test_y)
```

## Support Vector Regression

```python
from sklearn.svm import SVR

clf = SVR(gamma='scale', C=1.0, epsilon=0.2)
clf.fit(train_x, train_y)

test_y_svr = clf.predict(test_y)

np.sqrt(mean_squared_error(test_y_svr, test_y)), np.mean(test_y)
```
